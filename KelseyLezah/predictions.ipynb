```python
prediction_of_test = regr.predict(testing_data)
```

```python
testing_test = list(testing_test)
```

```python
sum = 0
rows = len(testing_test)
for i in range(rows):
    test = testing_test[i]
    prediction = prediction_of_test[i]
    diff = (test - prediction) ** 2
    sum = sum + diff
mse = sum / rows
print("MSE", mse)
```

```python
sum = 0
rows = len(testing_test)
for i in range(rows):
    test = testing_test[i]
    prediction = prediction_of_test[i]
    diff = abs(test - prediction)
    sum = sum + diff
mae = sum / rows
print("MAE", mae)
```
